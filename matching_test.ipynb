```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#===================
#分類標籤
labels = np.array(["面向1".decode("utf8"),"面向2".decode("utf8"),"面向3".decode("utf8"),
                   "面向4".decode("utf8"),"面向5".decode("utf8"),"面向6".decode("utf8")])
#6類
dataLenth = 6
#數據
data = np.array([1,4,3,6,4,8])
#====================

angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
data = np.concatenate((data, [data[0]])) 
angles = np.concatenate((angles, [angles[0]])) 

fig = plt.figure()
ax = fig.add_subplot(111, polar=True) # polar參數
ax.plot(angles, data, 'bo-', linewidth=2) 
ax.fill(angles, data, facecolor='#0000ff', alpha=0.25) 
ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
ax.set_title("求才分析".decode("utf8"), va='bottom', fontproperties="SimHei")
ax.set_rlim(0,10)
ax.grid(True)
plt.show()
```

```python
#定義成方法
def drawphase(data,name,color):
    import numpy as np
    import matplotlib.pyplot as plt
    %matplotlib inline
    labels = np.array(["面向1".decode("utf8"),"面向2".decode("utf8"),"面向3".decode("utf8"),
                       "面向4".decode("utf8"),"面向5".decode("utf8"),"面向6".decode("utf8")])
    dataLenth = 6
#    data = np.array([a,b,c,d,e,f])

    angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
    data = np.concatenate((data, [data[0]])) 
    angles = np.concatenate((angles, [angles[0]])) 

    fig = plt.figure()
    ax = fig.add_subplot(111, polar=True) # polar參數
    ax.plot(angles, data, 'bo-', linewidth=2) 
    ax.fill(angles, data, facecolor=color, alpha=0.25) 
    ax.set_thetagrids(angles * 180/np.pi, labels, fontproperties="SimHei")
    ax.set_title(name.decode("utf8"), va='bottom', fontproperties="SimHei")
    ax.set_rlim(0,10)
    ax.grid(True)
    plt.show()
```

```python
#Cosine Similarity 定義成方法

def matching(dataSetI,dataSetII):
    from scipy import spatial
    result = 1 - spatial.distance.cosine(dataSetI, dataSetII)
    return result
```

```python
#example 
import pandas as pd

#假設有4個求才需要:
data_need1 = [3,4,5,7,2,4]
data_need2 = [1,2,2,3,4,1]
data_need3 = [6,2,7,1,5,7]
data_need4 = [6,6,6,7,2,5]

#應徵者資料:
data_user = [4,4,5,7,1,3]

#相似度計算:
result1 = matching(data_user,data_need1)
result2 = matching(data_user,data_need2)
result3 = matching(data_user,data_need3)
result4 = matching(data_user,data_need4)
result_item = []
result_item.append({"Similarity1":result1,"Similarity2":result2,"Similarity3":result3,"Similarity4":result4})
df = pd.DataFrame(result_item)

#output
print "The matching results:"
df
```
